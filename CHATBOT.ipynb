```python
import nltk
from nltk.chat.util import Chat, reflections

# Define additional reflections for frequently asked questions
additional_reflections = {
    "how do I book a room": "You can book a room by providing your check-in and check-out dates.",
    "what are your room rates": "Our room rates vary depending on the type of room and the season.",
    "how can I cancel a reservation": "You can cancel a reservation by contacting our reservations department.",
    "do you offer discounts": "We offer discounts for group bookings and loyalty program members.",
    "breakfast timing": "Our breakfast is served from 7:00 AM to 10:00 AM.",
    "lunch timing": "Lunch is available from 12:00 PM to 2:00 PM.",
    "dinner timing": "Dinner is served from 6:00 PM to 9:00 PM.",
    "breakfast menu": "For breakfast, we offer a selection of items including pancakes, omelettes, and coffee.",
    "lunch menu": "Our lunch menu features a variety of sandwiches, salads, and daily specials.",
    "dinner menu": "Dinner options include steak, seafood, vegetarian dishes, and more.",
    "call for helper": "You can call for a helper by simply saying 'I need assistance' or 'Call for help.'",
    "fetch a helper": "To fetch a helper, approach any hotel staff member, or use our in-room phone to request assistance.",
}

# Combine the original and additional reflections
reflections = {**reflections, **additional_reflections}

# Define pattern-response pairs for the chatbot
pairs = [
    [
        r"hi|hello|hey",
        ["Hello! Welcome to our hotel. How may I assist you today?", "Hi there! How can I help you today?"]
    ],
    [
        r"book a room|room booking",
        ["Certainly, I can help you with that. Please provide your check-in and check-out dates."]
    ],
    [
        r"(\d{2}-\d{2}-\d{4}) (\d{2}-\d{2}-\d{4})",
        ["Great! I have rooms available from {} to {}. How many guests will be staying?".format("{0}", "{1}")]
    ],
    [
        r"(\d+)",
        ["Perfect! How many adults and children will be staying in the room?"]
    ],
    [
        r"adults (\d+) children (\d+)",
        ["Got it! Please provide your contact information to confirm the booking."]
    ],
    [
        r"reserve a table|restaurant reservation",
        ["Certainly! How many people will be dining with us?"]
    ],
    [
        r"(\d+)",
        ["Great! What date and time would you like to make the reservation?"]
    ],
    [
        r"how do I book a room",
        ["To book a room, please provide your check-in and check-out dates."],
    ],
    [
        r"what are your room rates",
        ["Our room rates vary depending on the type of room and the season. Please let us know your preferences."],
    ],
    [
        r"how can I cancel a reservation",
        ["You can cancel a reservation by contacting our reservations department. We can assist you with that."],
    ],
    [
        r"do you offer discounts",
        ["We offer discounts for group bookings and loyalty program members. Can I assist you with a booking or another inquiry?"],
    ],
    [
        r"breakfast timing",
        ["Our breakfast is served from 7:00 AM to 10:00 AM."],
    ],
    [
        r"lunch timing",
        ["Lunch is available from 12:00 PM to 2:00 PM."],
    ],
    [
        r"dinner timing",
        ["Dinner is served from 6:00 PM to 9:00 PM."],
    ],
    [
        r"breakfast menu",
        ["For breakfast, we offer a selection of items including pancakes, omelettes, and coffee."],
    ],
    [
        r"lunch menu",
        ["Our lunch menu features a variety of sandwiches, salads, and daily specials."],
    ],
    [
        r"dinner menu",
        ["Dinner options include steak, seafood, vegetarian dishes, and more."],
    ],
    [
        r"call for helper|fetch a helper",
        ["You can call for a helper by simply saying 'I need assistance' or 'Call for help.' To fetch a helper, approach any hotel staff member, or use our in-room phone to request assistance."],
    ],
    [
        r"(.*)",
        ["I'm here to assist you. If you have any other questions or specific requests, feel free to ask."]
    ]
]

# Create a chatbot instance
chatbot = Chat(pairs, reflections)

# Start the conversation
print("Hello, I'm your hotel assistant. How can I help you today?")

# Interactive chat loop
chatbot.converse()
```
